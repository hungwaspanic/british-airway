```python
texts = []

for i in range(1,3):
    URL = f"https://www.airlinequality.com/airline-reviews/british-airways/page/{i}/"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find(id="main")
    texts.extend(results.find_all("article", itemprop="review"))
texts

```

```python
name = []

final_date = []
date = []

customer_country = []

a_count = []
verified = []

review = []

type = []

seat = []

route = []

final_flown_date = []
date_flown = []

seat_review = []
seat_next_sibling = []
seat_count = []

cabin_review = []
cabin_next_sibling = []
cabin_count = []

food_review = []
food_next_sibling = []
food_count = []

ent_review = []
ent_next_sibling = []
ent_count = []

ground_review = []
ground_next_sibling = []
ground_count = []

wifi_review = []
wifi_next_sibling = []
wifi_count = []

value_review = []
value_next_sibling = []
value_count = []

rec_review = []
rec_next_sibling = []

    # Get review
    # a = text.find("h2", class_="text_header")

    # Get the name

for text in texts:
        name.append(text.find("span", itemprop="name").text.strip())


    # Get Date

for text in texts:
        date.append(text.find("time", itemprop="datePublished").text.strip())
for i in range(len(date)):
        date[i] = date[i].replace('st', '').replace('nd', '').replace('rd', '').replace('th', '') # Remove suffix
        date_obj = datetime.strptime(date[i], "%d %B %Y")
        encoded_date = date_obj.strftime("%Y-%m-%d")
        final_date.append(encoded_date)



    # Get country of customer
for text in texts:
        customer_country.append(text.find(text=re.compile('\((.*?)\)')))
    
    # Get verified

for text in texts:
        a_count.append(text.find("div", itemprop="reviewBody").text.strip())
for i in range(len(date)):
        if "trip verified" in a_count[i].lower():
            verified.append("verified")
        else:
            verified.append("not verified")

    # Get overall overview

for text in texts:
        review.append(text.find("span", itemprop="ratingValue").text.strip())

    # Get type of traveller

for text in texts:
        type.append(text.find("td", class_="review-rating-header type_of_traveller").find_next_sibling("td", class_="review-value").text.strip())

    # Get seat type

for text in texts:
        seat.append(text.find("td", class_="review-rating-header cabin_flown").find_next_sibling("td", class_="review-value").text.strip())
    
    # Get route

for text in texts:
        route.append(text.find("td", class_="review-rating-header route").find_next_sibling("td", class_="review-value").text.strip())
    
    # Get date flown

for text in texts:
        date_flown.append(text.find("td", class_="review-rating-header date_flown").find_next_sibling("td", class_="review-value").text.strip())
for i in range(len(date)):
        final_flown_date.append(datetime.strptime(date_flown[i], "%B %Y").strftime("%Y-%m"))    

    # seat_comfort review

for text in texts:
        seat_review.append(text.find("td", class_="review-rating-header seat_comfort"))
for i in range(len(date)):
        if seat_review[i] is not None:
            seat_next_sibling.append(seat_review[i].find_next_sibling("td", class_="review-rating-stars stars"))
        else:
            seat_next_sibling.append("None")
for i in range(len(date)):
        seat_count.append(str(seat_next_sibling[i]).count("star fill"))


    # 11, cabin staff service review

for text in texts:
        cabin_review.append(text.find("td", class_="review-rating-header cabin_staff_service"))
for i in range(len(date)):
        if cabin_review[i] is not None:
            cabin_next_sibling.append(cabin_review[i].find_next_sibling("td", class_="review-rating-stars stars"))
        else:
            cabin_next_sibling.append("None")
for i in range(len(date)):
        cabin_count.append(str(cabin_next_sibling[i]).count("star fill"))


    # 12, food&beverages review

for text in texts:
        food_review.append(text.find("td", class_="review-rating-header food_and_beverages"))
for i in range(len(date)):
        if food_review[i] is not None:
            food_next_sibling.append(food_review[i].find_next_sibling("td", class_="review-rating-stars stars"))
        else:
            food_next_sibling.append("None")
for i in range(len(date)):
        food_count.append(str(food_next_sibling[i]).count("star fill"))
    

    # 13, inflight entertainment review

for text in texts:
        ent_review.append(text.find("td", class_="review-rating-header inflight_entertainment"))
for i in range(len(date)):
        if ent_review[i] is not None:
            ent_next_sibling.append(ent_review[i].find_next_sibling("td", class_="review-rating-stars stars"))
        else:
            ent_next_sibling.append("None")
for i in range(len(date)):
        ent_count.append(str(ent_next_sibling[i]).count("star fill"))

    
    # 14, ground service review

for text in texts:
        ground_review.append(text.find("td", class_="review-rating-header ground_service"))
for i in range(len(date)):
        if ground_review[i] is not None:
            ground_next_sibling.append(ground_review[i].find_next_sibling("td", class_="review-rating-stars stars"))
        else:
            ground_next_sibling.append("None")
for i in range(len(date)):
        ground_count.append(str(ground_next_sibling[i]).count("star fill"))

    # 15, wifi review

for text in texts:
        wifi_review.append(text.find("td", class_="review-rating-header wifi_and_connectivity"))
for i in range(len(date)):
        if wifi_review[i] is not None:
            wifi_next_sibling.append(wifi_review[i].find_next_sibling("td", class_="review-rating-stars stars"))
        else:
            wifi_next_sibling.append("None")
for i in range(len(date)):
        wifi_count.append(str(wifi_next_sibling[i]).count("star fill"))

    # 16, value for money review

for text in texts:
        value_review.append(text.find("td", class_="review-rating-header value_for_money"))
for i in range(len(date)):
        if value_review[i] is not None:
            value_next_sibling.append(value_review[i].find_next_sibling("td", class_="review-rating-stars stars"))
        else:
            value_next_sibling.append("None")
for i in range(len(date)):
        value_count.append(str(value_next_sibling[i]).count("star fill"))

    # 17, Recommended or not

for text in texts:
        rec_review.append(text.find("td", class_="review-rating-header recommended"))
for i in range(len(date)):
        rec_next_sibling.append(rec_review[i].find_next_sibling("td", class_="review-value").text)

```

```python
result = []
for x in range(len(final_date)):
    result.append(
        {"Name": name[x],
         "review_date": final_date[x],
         "location": customer_country[x],
         "trip_verified": verified[x],
         "overall_review": review[x],
         "type_of_traveller": type[x],
         "seat_type": seat[x],
         "route": route[x],
         "date_flown": date_flown[x],
         "seat_comfort_review": seat_count[x],
         "cabin_service_review": cabin_count[x],
         "food_and_beverage_review": food_count[x],
         "inflight_entertainment": ent_count[x],
         "ground_service": ground_count[x],
         "wifi_review": wifi_count[x],
         "value_for_money": value_count[x],
         "recommended": rec_next_sibling[x]
         }
    )
```

```python
df = pd.DataFrame(result)
```
